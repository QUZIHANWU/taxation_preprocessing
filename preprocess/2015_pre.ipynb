### 3：剔除管理层平均工资比普通员工平均工资低的公司,因为这类公司披露的高管薪酬可能仅是津贴,或者数据错误：

- 管理层平均工资 = 管理层年度薪酬总数/（非独立董事人数+公司监事+高管人数-UnpaidNumber）
- 普通员工平均工资 = （支付给职工以及为职工支付的现金-管理层年度薪酬总数）/ （员工总数-（非独立董事+公司监事+高管人数））

```python
full_data.columns
```

```python
full_data['人数']=full_data['公司监事'].apply(lambda x: len(x.split(',')))
```

```python
full_data.head(1)
```

```python
full_data.isnull().sum()
```

```python
check = full_data['非独立董事人数'] + full_data['人数'] + full_data['高管人数'] - full_data['UnpaidNumber']
```

```python
check_data = pd.DataFrame(check, columns=['check_0'])
check_data.to_excel('/Users/quzihanwu/Desktop/Preprocessing_copy_by_Martin/data/raw/2015/check_0.xlsx')
```

```python
check_data[(check_data['check_0'] == 0)]
```

```python
full_data['check_0'] = check_data
full_data
```

```python
full_data = full_data[(full_data['check_0'] != 0)]
```

```python
full_data['管理层平均工资'] = full_data['管理层年度薪酬总额'] / (
   full_data['非独立董事人数'] + full_data['人数'] + full_data['高管人数'] 
      - full_data['UnpaidNumber']
) *10000
full_data.head(1)
```

```python
full_data['普通员工平均工资'] = (full_data ['支付给职工以及为职工支付的现金'] 
   - full_data['管理层年度薪酬总额']) / (
      full_data['员工总数'] - (
         full_data['非独立董事人数'] + full_data['人数'] + full_data['高管人数']
      )
) *10000
full_data.head(1)
```

```python
full_data = full_data[(full_data["管理层平均工资"] > full_data["普通员工平均工资"])]
```

### .to_table

```python
import math

def get_log(full_data):
   result_log = []
   list_log = full_data.values
   for i in list_log:
      result_log_n = math.log(i)
      result_log.append(result_log_n)
   return result_log

list_result = get_log(full_data['管理层平均工资'])
full_data['lnAMW'] = list_result

list_result = get_log(full_data['普通员工平均工资'])
full_data['lnAEW'] = list_result

full_data['收入不平等'] = full_data['管理层平均工资'] / full_data['普通员工平均工资']
list_result = get_log(full_data['收入不平等'])
full_data['lnWFI'] = list_result

full_data['净利润'] = full_data['净利润']*10000
full_data['递延所得税负债增加'] = full_data['递延所得税负债增加']*10000
full_data['递延所得税资产减少'] = full_data['递延所得税资产减少']*10000
full_data['ETR'] = full_data['所得税费用'] / (
   (
      full_data['所得税费用'] + full_data['净利润'] - (
         full_data['递延所得税负债增加'] + full_data['递延所得税资产减少']
      )/ (full_data['年末所得税率'])
   )
)

list_result = get_log(full_data['资产总计']*10000)
full_data['lnSize'] = list_result

full_data['OCF'] = full_data['经营活动产生的现金流量净额'] / full_data['资产总计']

full_data['Lev'] = full_data['资产负债率']/100

full_data['Growth'] = full_data['营业收入']/100

full_data['Top1'] = full_data['大股东持股比例']

full_data['indepent'] = (full_data['独立董事人数'] / full_data['董事会人数'])*100

full_data['MH'] = (full_data['监管层持股数'] / full_data['总股数'])*100

full_data["Dual"] = np.where(full_data["董事长和总经理是否同一人"] =='Y', "1", "0")

full_data['国有判断'] = full_data['公司属性'].str[2:4]
full_data['SOE'] = np.where(full_data['国有判断'] =='国有', "1", "0")
```
